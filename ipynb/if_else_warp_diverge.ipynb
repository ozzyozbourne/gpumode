```python
%%writefile warp_divergence.cu
#include <stdio.h>

__global__ void warpDivergeExample() {
    int tid = threadIdx.x;

    // This if-else causes warp divergence
    if (tid < 16) {
        // First 16 threads (0-15) execute Path A
        tid = tid * 2;
    } else {
        // Last 16 threads (16-31) execute Path B
        tid = tid * 3;
    }
}

int main() {
    // Launch with 1 block, 32 threads (1 warp)
    warpDivergeExample<<<1, 32>>>();
    cudaDeviceSynchronize();

    printf("Kernel executed - check cuda-gdb to see divergence\n");
    return 0;
}
```
