```python
%%writefile warp_divergence.cu
#include <stdio.h>

__global__ void warpDivergeExample() {
    int tid = threadIdx.x;

    if (tid < 32) {
      tid = tid * 2;
    }
}

int main(void) {

    warpDivergeExample<<<1, 32>>>();
    cudaDeviceSynchronize();

    return 0;
}
```
