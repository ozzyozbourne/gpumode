```julia
using CUDA, Printf

# Define a proper vector-add kernel: C = A + B
function vector_add!(A, B, C, N)
    idx = threadIdx().x + (blockIdx().x - 1) * blockDim().x
    @inbounds C[idx] = A[idx] + B[idx]
    return
end

# Create input & output CuArrays along with Compute execution configuration
N, blocks = 32, 1
A, B, C = CUDA.fill(1.0f0, N), CUDA.fill(2.0f0, N), similar(CUDA.fill(1.0f0, N))

# Launch the kernel
@cuda threads=N blocks=blocks vector_add!(A, B, C, N)

# Copy back to CPU memory
h_C = Array(C)

for (i, val) in enumerate(h_C)
    @printf("C[%2d] = %.2f\n", i, val)
end
```
